```python
df.drop(['Startup Name','Year Founded'],inplace = True,axis = 'columns')
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
le = OneHotEncoder(drop="first",sparse_output=False,handle_unknown='ignore')
encoded = le.fit_transform(df[['Industry','Country']])
encoded_df = pd.DataFrame(encoded,columns=le.get_feature_names_out(['Industry','Country']))
encoded_df.head()
```

```python
final_df = pd.concat([df,encoded_df],axis = 1)
final_df.head()
final_df.drop(['Industry','Country'],axis = 1, inplace=True)
```

```python
final_df = final_df.applymap(lambda x: f"{x:.0f}" if isinstance(x, (int, float)) else x)
```

```python
final_df.head()
```

```python
X = final_df.drop('Valuation (USD)', axis = 1)
Y = final_df['Valuation (USD)']
```

```python
from sklearn.model_selection import train_test_split , cross_val_score
X_train , X_test , y_train , y_test = train_test_split(X,Y,train_size= 0.8, random_state=41)
```

```python
numerical = ['Funding Rounds', 'Investment Amount (USD)', 'Number of Investors','Growth Rate (%)', 'Company Age']
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train[numerical] = ss.fit_transform(X_train[numerical])
X_test[numerical] = ss.transform(X_test[numerical])
```

```python
pd.DataFrame(X_train).head()
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,y_train)
y_predict = LR.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2score=r2_score(y_test,y_predict)
r2score
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
import numpy as np
ridge  = Ridge(max_iter=10000)
param_grid = {
    'alpha': np.logspace(-4,4,50)
}

# Define GridSearchCV
grid = GridSearchCV(estimator=ridge,
                    param_grid=param_grid,
                    scoring='r2',
                    cv=5,
                    n_jobs=-1)

# Fit on training data
grid.fit(X_train, y_train)

# Best parameters and score
print("Best alpha:", grid.best_params_)
print("Best RÂ² Score (CV):", grid.best_score_)

# Evaluate on test set
y_pred = grid.best_estimator_.predict(X_test)
print("Test RÂ² Score:", r2_score(y_test, y_pred))
```

```python
r2score=r2_score(y_test,y_predict)
r2score
```

```python
import warnings
warnings.filterwarnings('ignore')
cross_val_score(grid.best_estimator_,X,Y,cv=5)
```

```python
best_model = grid.best_estimator_
```

```python
import joblib
joblib.dump(best_model,open('model.pkl','wb'))
joblib.dump(le,open('le.pkl','wb'))
joblib.dump(ss,open('ss.pkl','wb'))
```
